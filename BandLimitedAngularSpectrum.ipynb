```python
import numpy as np
import cv2 as cv
import torch
import matplotlib.pyplot as plt
import torchvision.io
from torchvision.transforms import transforms
from PIL import Image
import math
from math import e
```

```python
class Light():
  def __init__(self,height,width,wave_lenght):
    self.height = height
    self.width = width
    self.wave_lenght = wave_lenght
    self.g = numpy.ones((height,width))

  def forward(self,z):
    r = (self.height**2+self.width**2+z**2)**1/2
    h = (math.e**(1j*2*math.pi*r*(self.wave_lenght)**-1)*z/r**2)*(1/2*math.pi*r+1/1j*self.wave_lenght)
    g = self.g * h 

    return g
```
